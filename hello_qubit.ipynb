# Hello Qubit - My First Quantum Circuit

This notebook builds and runs the simplest possible quantum program:
putting a single qubit into superposition, then measuring it.

**What you'll see:** a qubit that starts as a definite 0, gets put into
superposition (50/50 chance of being 0 or 1), and is then measured -
demonstrated by running the circuit 1000 times and plotting the results.

```python
# QuantumCircuit: the core object used to build quantum programs in Qiskit
from qiskit import QuantumCircuit

# AerSimulator: a fast, local simulator for running circuits without real hardware
from qiskit_aer import AerSimulator

# plot_histogram: visualizes measurement results as a bar chart
from qiskit.visualization import plot_histogram
```

## Step 1: Build the circuit

A qubit always starts in a definite state, |0⟩ - like a coin lying flat,
not yet flipped.

The **Hadamard gate (H)** puts the qubit into **superposition** - after
this gate, the qubit isn't "0" or "1", it's in a state where measuring it
gives a 50/50 chance of either outcome.

Measurement is what forces the qubit to "collapse" into one definite value.

```python
# Create a circuit with 1 qubit and 1 classical bit
# (the classical bit is where the measurement result gets stored)
qc = QuantumCircuit(1, 1)

# Apply a Hadamard gate to qubit 0 — this creates superposition
qc.h(0)

# Measure qubit 0, store the result in classical bit 0
qc.measure(0, 0)

# Draw the circuit as a diagram
qc.draw("mpl")
```

## Step 2: Run it many times and look at the results

Since a single measurement only gives one random outcome, we run the
circuit many times ("shots") to see the underlying probability.
Because of the Hadamard gate, we expect roughly 50% "0" and 50% "1" -
not exactly 50/50, since quantum measurement is genuinely random.

```python
# Create a local simulator
sim = AerSimulator()

# Run the circuit 1000 times ("shots") on the simulator
job = sim.run(qc, shots=1000)

# Get the results of the job
result = job.result()

# Get the measurement counts — how many times we got '0' vs '1'
counts = result.get_counts()

# Print the raw counts
print(counts)

# Plot the counts as a bar chart
plot_histogram(counts)
```

## What I learned

- A qubit starts in a definite state (0), and the Hadamard gate puts it
  into superposition.
- Measuring a qubit in superposition gives a random 0 or 1, roughly 50/50
  over many runs.
