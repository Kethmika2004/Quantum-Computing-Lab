# 🔗 Bell States - My First Entangled Qubits

This notebook builds a **Bell state** - two qubits that become
**entangled** 🌀, meaning their measurement outcomes are correlated
no matter how far apart they are.

This is the first genuinely "quantum" phenomenon with no classical
equivalent - it's the building block behind teleportation 🚀, many
quantum algorithms, and quantum cryptography 🔐.

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

## 🛠️ Step 1: Build the circuit

Two qubits, both starting at |0⟩.

- A **Hadamard gate (H)** on qubit 0 puts it into superposition 🎲
- A **CNOT gate (CX)** then entangles qubit 1 with qubit 0 - it flips
  qubit 1 *only if* qubit 0 is 1

The result: measuring either qubit is still random (50/50), but the
two outcomes will always **match** - both 0, or both 1. Never mixed.
That correlation, even though each individual outcome is random, is entanglement ✨.

```python
# 2 qubits, 2 classical bits
qc = QuantumCircuit(2, 2)

# Put qubit 0 into superposition
qc.h(0)

# Entangle qubit 1 with qubit 0 (CNOT: flip qubit 1 if qubit 0 is 1)
qc.cx(0, 1)

# Measure both qubits
qc.measure([0, 1], [0, 1])

qc.draw("mpl")
```

## ▶️ Step 2: Run it and check the correlation

If entanglement is working, we should only ever see `00` or `11`
in the results - never `01` or `10` 🚫.

```python
sim = AerSimulator()
job = sim.run(qc, shots=1000)
result = job.result()
counts = result.get_counts()

print(counts)
plot_histogram(counts)
```

## 📝 What I learned

- Entanglement means two qubits' measurement outcomes are correlated,
  even though each one individually is still random 🎯
- I only saw `00` and `11` outcomes - never `01` or `10` - which is
  the signature of this specific Bell state 🔍
